```python
# Imports
import pandas as pd
#import numpy as np
```

### 1. Import all datasets

```python
json_path_eixample = "../data/eixample/dades_netes.json"
json_path_gracia = "../data/gracia/dades_netes.json"
json_path_santandreu = "../data/sant andreu/dades_netes.json"
json_path_santmarti = "../data/sant marti/dades_netes.json"
json_path_sants = "../data/sants/dades_netes.json"
json_path_sarria = "../data/sarria/dades_netes.json"

json_path_hospitalet = "../data/hospitalet/dades_netes.json"
```

```python
input_eixample = pd.read_json(json_path_eixample)
input_gracia = pd.read_json(json_path_gracia)
input_santandreu = pd.read_json(json_path_santandreu)
input_santmarti = pd.read_json(json_path_santmarti)
input_sants = pd.read_json(json_path_sants)
input_sarria = pd.read_json(json_path_sarria)

input_hospitalet = pd.read_json(json_path_hospitalet)

#input_eixample.dtypes
```

### 2. Merge all datasets

```python
input_data = pd.concat([input_eixample, input_gracia, input_santandreu, input_santmarti, input_sants, input_sarria, input_hospitalet])

print(input_data.shape)
```

### 3. Data Processing

```python
df = input_data.copy()

df = df.reset_index(drop=True)
```

```python
# Drop duplicates
subset_columns = [col for col in df.columns if col not in ['schedule', 'name']]
df = df.drop_duplicates(subset=subset_columns)

# Sort by number of reviews
df = df.sort_values(by='num_reviews', ascending=False)
```

### 4. Export data

```python
# --- Save to CSV ---
output_path_csv = "../data/dades_finals.csv"
df.to_csv(output_path_csv, index=False)
```

```python
# --- Save to JSON ---
output_path_json = "../data/dades_finals.json"
df.to_json(output_path_json, orient='records')
```
