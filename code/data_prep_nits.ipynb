```python
# Imports
import pandas as pd
#import numpy as np
```

### 1. Import data and DF creation

```python
json_path = "../data/dades_amb_emails.json"

#input_csv = pd.read_csv(json_path, sep=",")
input_json = pd.read_json(json_path)

input_json.head(5)
#input_csv.head(5)

```

### 2. Prepare DataFrame

```python
main_df = input_json.copy()

main_df = main_df.reset_index(drop=True)
main_df.index += 1
#main_df.head()
```

```python
# Drop row if num_reviews or reviews_rating is "NOT_FOUND"
main_df = main_df[main_df['num_reviews'] != "NOT_FOUND"]
main_df = main_df[main_df['reviews_rating'] != "NOT_FOUND"]

# Change num_reviews and reviews_rating to int and float
main_df['num_reviews'] = main_df['num_reviews'].astype(int)
main_df['reviews_rating'] = main_df['reviews_rating'].astype(float)
```

```python
# Sort by number of reviews
main_df = main_df.sort_values(by='num_reviews', ascending=False)
```

```python
# Rename auxiliar column
main_df = main_df.rename(columns={"auxiliar": "auxiliar(borrar)"})

# Reorder columns
main_df = main_df.iloc[:, [0, 1, 5, 2, 3, 11, 4, 7, 8, 9, 10, 6]]
```

```python
# Selecciona las columnas a considerar (excluyendo 'schedule' y 'name')
subset_columns = [col for col in main_df.columns if col not in ['schedule', 'name']]

# Elimina duplicados solo usando las columnas seleccionadas
main_df = main_df.drop_duplicates(subset=subset_columns)
```

### 3. Export DataFrame to CSV and JSON

```python
df = main_df.copy()
```

```python
# --- Save to CSV ---
output_path_csv = "../data/dades_netes.csv"
df.to_csv(output_path_csv, index=False)
```

```python
# --- Save to JSON ---
output_path_json = "../data/dades_netes.json"
df.to_json(output_path_json, index=False)
```
